```python
# cleaning the Data by renaming the columnn names
df=df.rename(columns={'Province/State':'State',
                     'Country/Region':'Country',
                     'ObservationDate':'Date'})
df
```

```python
#changing the country and date time format
df.loc[df['Country']=='Mainland China','Country']='China'
df['Date'] = pd.to_datetime(df['Date'],format='%m/%d/%Y')
df['Day'] = df.Date.dt.dayofyear
```

```python
df
```

```python
#filling the null values
df['State'].fillna(method='ffill',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull().any()
```

```python
#checking the last update
df['Last Update']=pd.to_datetime(df['Last Update'])
df
```

```python
# changing values float to int
```

```python
df['Confirmed'] = df['Confirmed'].astype('int64')
df['Deaths'] = df['Deaths'].astype('int64')
df['Recovered'] = df['Recovered'].astype('int64')
df
```
